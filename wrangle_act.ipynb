<a id='api-call'></a>
##### Retrieve tweeter data via API call

````python
# Get tweet JSON data via tweed ID using Tweepy library
# Using parameter `parser=tweepy.parsers.JSONParser()` we can parse json file directly.
count = 0
tweet = []
fails_dict = {}

for tweet_id in tweet_Archive['tweet_id']:
    count +=1
    print(str(count) + ": " + str(tweet_id))
    try:
        start = time.time()
        data = api.get_status(tweet_id, tweet_mode='extended', wait_on_rate_limit=True, wait_on_rate_limit_notify=True, parser=tweepy.parsers.JSONParser())
        tweet.append(data)
        end = time.time()
        print('Success')
        print("query time", end-start)
    except tweepy.TweepError as e:
        print(e)
        print("Fail")
        fails_dict[tweet_id] = e
    except:
        print(e)       
````

#### Tidiness

***Rules for tidy data:***
```
Each variable forms a column
Each observation forms a row
Each type of observational unit forms a table.
(Source: Udacity Knowledge and Lectures in Data Wrangling Lesson)
```
***Another helpful guiadance for better udenrstanding:***
```
Column headers are values, not variable names.
Multiple variables are stored in one column.
Variables are stored in both rows and columns.
Multiple types of observational units are stored in the same table.
A single observational unit is stored in multiple tables.
source (Udacity Knowledge)
```

##### Issue name
***`twitter_archive` table: One variable is spread across four different columns: doggo, floofer, pupper, puppo columns<br>
`twitter_archive` table: Some dogs have multiple stages***

##### Define
*These four columns (doggo, floofer, pupper, puppo) should be combined into a single column named `dog_stage`.
*Multiple dog stages should be neatly formatted, sepatated with comma.*

##### Code

````python
# Replace None in stage columns with empty string.
twitter_archive_clean.doggo.replace('None', '', inplace=True)
twitter_archive_clean.floofer.replace('None', '', inplace=True)
twitter_archive_clean.pupper.replace('None', '', inplace=True)
twitter_archive_clean.puppo.replace('None', '', inplace=True)

# Combine stage columns
twitter_archive_clean['stage'] = twitter_archive_clean.doggo + twitter_archive_clean.floofer + twitter_archive_clean.pupper + twitter_archive_clean.puppo

# Format entries with multiple dog stages
twitter_archive_clean.loc[twitter_archive_clean.stage == 'doggopupper', 'stage'] = 'doggo,pupper'
twitter_archive_clean.loc[twitter_archive_clean.stage == 'doggopuppo', 'stage'] = 'doggo,puppo'
twitter_archive_clean.loc[twitter_archive_clean.stage == 'doggofloofer', 'stage'] = 'doggo,floofer'
````

````python
# Value counts
twitter_archive_clean.stage.value_counts()
````

````python
# Fill empty string with NaNs
twitter_archive_clean.loc[twitter_archive_clean.stage == '', 'stage'] = np.nan
````

````python
# Drop unneccesary columns
twitter_archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
````

````python
# Set the index to DatetimeIndex in order to use resample method
datedf_set_index = master_df.set_index(pd.DatetimeIndex(master_df['date']))
````
